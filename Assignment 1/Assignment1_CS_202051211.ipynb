```python
import pandas as pd
import json
```

### 1. Create two data frames by reading above two files.

```python
district_df = pd.read_csv('district_level_service_msme.xls')
district_df.head(5)
```

```python
filepath="district_level_manufacturing_msme.json"
with open(filepath, 'r') as f:
  data = json.load(f)
  district_manufacturing_df = pd.DataFrame(data["data"],)

district_manufacturing_df.head()
```

```python
district_manufacturing_df.columns = district_df.columns

district_df.info()

district_manufacturing_df.head()
district_manufacturing_df.info()
```

### 2. Find out total ”Small” Manufacturing MSME in India.

```python
district_df['SMALL'].sum()
```

### 3. Create a dataframe having state wise total number of ”Micro”,”Small” and ”Medium” Services MSE (as shown below) and save the results as a CSV file.

```python
cols = ['MICRO','SMALL','MEDIUM','Total','Lg_Dist_Code']
district_manufacturing_df[cols] = district_manufacturing_df[cols].astype(int)

temp_df = district_df.groupby('STATE_NAME').sum()[['MICRO','SMALL','MEDIUM']]
temp_df.head(5)

temp_df.to_csv("state_msme.csv")

columns = ['STATE_NAME','DISTRICT_NAME','Lg_Dist_Code','Last_Updated']
result_df = pd.merge(left=district_manufacturing_df,right=district_df,on=columns)
result_df.head()
```

### 4. Join the both the data frame based on common STATE NAME, DISTRICT NAME, Lg Dist Code and Last Updated. The result should look like below. ”x” and ”y” in below image represent the manufacturing MSME and service MSME respectively.

```python
columns = ['STATE_NAME','DISTRICT_NAME','Lg_Dist_Code','Last_Updated']
result_df = pd.merge(left=district_manufacturing_df,right=district_df,on=columns)
result_df.head()
```

### 5. Create a Pivot Table having rows STATE NAME and columns Service and Manufacturing ”MSME” as show in below. Use ”Sum” to add up district wise number.

```python
pivot_msme = pd.pivot_table(result_df,index=['STATE_NAME'],aggfunc=sum)
pivot_msme= pivot_msme.drop(['Lg_Dist_Code','Total_x','Total_y'],axis=1)
pivot_msme.head()
```
