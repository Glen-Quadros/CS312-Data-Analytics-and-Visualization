```python
countries = ['India', 'Japan', 'China', 'United Kingdom', 'United States']
filtered_df = co2_df[(co2_df['Country Name'].isin(countries))]

year_2012 = filtered_df.groupby('Country Name')['2012'].sum()
year_2013 = filtered_df.groupby('Country Name')['2013'].sum()
year_2014 = filtered_df.groupby('Country Name')['2014'].sum()
year_2015 = filtered_df.groupby('Country Name')['2015'].sum()

print(np.sum(year_2012))
print(np.sum(year_2013))
print(np.sum(year_2014))
print(np.sum(year_2015))

year_2015
```
